# 3º Exercício: String

##**Parte 1**

Aplique três métodos distintos na string abaixo.

In [ ]: cancao = 'Roda mundo, roda gigante, roda moinho, roda pião.'

```python
canção = 'Roda mundo, roda gigante, roda moinho, roda pião'

# maiusculo: string.upper()
print(canção.upper())
```

```python
#posicao: string.replace(antigo,novo)
posicao = canção.find('pião')
print(posicao)
```

```python
# substituição: string.replace(antigo, novo)
print(canção.replace('pião','escorpião'))
```

##**Parte 2**

Extraia da string abaixo o valor da taxa selic na variável selic e o valor do ano na variável ano.

Imprima os valores na tela.



```python
noticia = 'Selic vai a 2,75% e supera expectativas; é a primeira alta em 6 anos.'
```

```python
texto = noticia[0:6]
print(texto)

indice_selic_inicio = noticia.find('Selic')
print(indice_selic_inicio)

indice_selic_final = noticia.find(' vai')
print(indice_selic_final)

selic = noticia[indice_selic_inicio:indice_selic_final]
print(selic)
```

```python
indice_anos_inicio = noticia.find('anos')
print(indice_anos_inicio)

indice_anos_final = noticia.find(' .')
print(indice_anos_final)

anos = noticia[indice_anos_inicio:indice_anos_final]
print(anos)
```
